# Exploratory Data Analysis (EDA)

In this notebook, we analyze the SKU110K / Open Images dataset properties to inform our object detection architecture choices. We examine image resolutions, bounding box distributions, and aspect ratios.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
import glob
import os
from pathlib import Path

sns.set_theme(style="whitegrid")
```

```python
data_dir = "../data/raw"
images = glob.glob(f"{data_dir}/*.jpg")
print(f"Total RAW images found: {len(images)}")
```

### 1. Image Resolution Analysis

Let's sample images to see the distribution of widths and heights. Dense shelves are heavily impacted by extreme aspect ratios or high compression during resize.

```python
# Analyze image dimensions
widths = []
heights = []

# Sample first 500 images for speed in EDA
for img_path in images[:500]:
    img = cv2.imread(img_path)
    if img is not None:
        h, w, _ = img.shape
        heights.append(h)
        widths.append(w)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(widths, heights, alpha=0.5, color='royalblue')
plt.title("Image Resolutions Distribution")
plt.xlabel("Width (pixels)")
plt.ylabel("Height (pixels)")
plt.show()
```

### Conclusion on Resolution
Retail datasets often come in standard smartphone formats (e.g. 1920x1080 HD, or 4K). For YOLOv8 training, resizing uniformly to `640x640` or `1280x1280` preserving the aspect ratio (via padding) gives the best inference speed vs accuracy ratio.
